```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Tour_Winners_data_1.csv')
df1 = pd.read_csv('Tour_Winners_data_2.csv')
df2 = pd.read_csv('Tour_Winners_data_3.csv')
df3 = pd.read_csv('TDF_Riders_History.csv')
```

```python
# Contar o número de vitórias por idade
age_counts = df['age'].value_counts()

# Ordenar os valores pelo índice (idade)
age_counts = age_counts.sort_index()

# Plotar o gráfico de barras usando Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=age_counts.index, y=age_counts.values, color='skyblue')
plt.title('Número de Vitórias no Tour de France por Idade do Vencedor')
plt.xlabel('Idade')
plt.ylabel('Número de Vitórias')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Agrupando os dados pelo peso do vencedor e contando o número de vitórias para cada peso
weight_counts = df['weight_(Kg)'].value_counts()

# Criando o gráfico de barras usando Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=weight_counts.index, y=weight_counts.values, color='skyblue')

plt.title('Número de Vitórias por Peso dos Vencedores do Tour de France')
plt.xlabel('Peso (Kg)')
plt.ylabel('Número de Vitórias')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=20, kde=True, color='skyblue')
plt.title('Distribuição da Idade dos Vencedores do Tour de France')
plt.xlabel('Idade')
plt.ylabel('Número de Vitórias')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Contagem de nacionalidades dos vencedores
country_counts = df['Country'].value_counts().head(5)  # Seleciona os 5 maiores vencedores

# Plotar o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct=lambda p: f'{int(p * sum(country_counts) / 100)}', startangle=140)
plt.title('Top 5 Países com Mais Vencedores do Tour de France')
plt.axis('equal')  # Garante que o gráfico seja circular
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Contagem de vitórias por vencedor
winner_counts = df['Winner'].value_counts().head(10)

# Plotagem do gráfico de barras usando Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=winner_counts.index, y=winner_counts.values, palette='viridis', hue=winner_counts.index)
plt.title('Top 10 Maiores Vencedores do Tour de France')
plt.xlabel('Vencedor')
plt.ylabel('Número de Vitórias')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df_filtered = df[(df['Year'] >= 2010) & (df['Year'] <= 2023)]

winner_counts = df_filtered['Winner'].value_counts().head(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=winner_counts.index, y=winner_counts.values, palette='viridis', hue=winner_counts.index)
plt.title('Top 5 Maiores Vencedores do Tour de France (2010-2023)')
plt.xlabel('Vencedor')
plt.ylabel('Número de Vitórias')
plt.xticks(rotation=45, ha='right')
plt.yticks(range(max(winner_counts.values) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
